- 生活中不乏这样的现象，一项技能或事物在开始时只有极少数人掌握并受益其中，其他人对这些技能或事物要么没有机会接触，要么心存误解，但随着社会发展和生活方式的变迁，它们逐渐扩散成为大众学习和掌握的基本技能，比如使用手机、网络、英语等等。这其中有一项，人们对其原理不清楚甚至有不少误解但每天受益其中，这就是编程。
- 和二三十年前的开车一样，现在从事编程的人叫程序员或戏称为码农，而我们每天浏览的网页、使用的APP、交通红绿灯、机器自动化运转等等，除了硬件，背后的一切都来自编程控制，完成人们预想达到的目标。大部分人只是享受着程序员带给他们的便利，自己却“独善其身”，无法创造自己想要的便利。如同小时候坐上汽车被司机带到目的地，不能按自己想要的路线前行，对于追求自由人这是无法忍受的事实。

## Python当歌
- 提到编程，你或许听说过C语言、汇编语言、Java和机器语言等等各种名词和术语，在正式学Python之前，先来理顺一下这些让你眼晕的名词。
### 机器语言
- 既然是给计算机编程，当然得知道它能识别的信号是什么？二进制是计算机唯一认识的信号。机器语言是用**二进制**代码表示的计算机能直接识别和执行的一种机器指令的集合。一条指令就是机器语言的一个语句，它是一组有意义的二进制代码。它是计算机的设计者通过计算机的硬件结构赋予计算机的操作功能。
- 机器语言具有灵活、直接执行和速度快等特点。不过不同型号的计算机其机器语言是不相通的，按着一种计算机的机器指令编制的程序，不能在另一种计算机上执行。用机器语言编写程序，编程人员要首先熟记所用计算机的全部指令代码和代码的涵义。编出的程序全是些0和1的指令代码，直观性差，还容易出错。除了计算机生产厂家的专业人员外，绝大多数的程序员已经不再去学习机器语言了。
- 所以对于机器语言，你只需要了解它的功能和优缺点就行，不需要去学习掌握它，仅此而已。
### 汇编语言
- 为了解决使用机器语言编写应用程序所带来的一系列问题，人们首先想到使用**助记符号**来代替不容易记忆的机器语言。这种助记符号来表示计算机指令的语言称为**符号**语言，也称汇编语言。在汇编语言中，每一条用符号来表示的汇编语言指令与计算机机器语言指令一一对应，记忆难度大大减少了，也易于检查和修改程序错误。
- 用汇编语言编写的程序称为源程序，计算机不能直接识别和处理源程序，必须通过某种方法将它翻译成为计算机能够理解并执行的机器语言，执行这个翻译工作的程序称为汇编程序。通过一个示例来理解汇编语言更直观：

```python
操作:将寄存器BX的内容送到AX中 
 
1000100111011000   机器语言
 
mov ax,bx          汇编语言
```

- 汇编程序就是执行将“mov ax,bx”翻译为机器能识别的“1000100111011000”工作。汇编语言是各种编程语言中与硬件关系最密切、最直接的一种,在时间和空间的效率上也最高的一种，它是高等院校计算机应用技术必修的专业课程之一，对于训练学生掌握程序设计技术，熟悉上机操作和程序调试技术有重要作用。但作为初学编程的人们，汇编语言还是显得低效、繁琐也很少接触，所以了解就行。
### 高级语言
- 高级语言是一类接近于人类的**自然语言和数学语言**的程序设计语言的统称。以人类的日常语言为基础，使用一般人易于接受的文字来表示（例如汉字、不规则英文或其他外语），从而使程序员编写更容易，亦有较高的可读性，以方便对电脑认知较浅的人亦可以大概明白其内容。由于早期电脑业的发展主要在美国，因此一般的高级语言都是以英语为蓝本。使用高级语言编写程序的优点是：编程相对简单、直观、易理解、不容易出错。我们常听说的**C语言、Basic、C++、Java、VB、Javascript**还有我们即将学习的**Python**，都属于高级语言。
- 高级语言是独立于计算机的，因而用高级语言编写的计算机程序通用性好，具有较好的移植性，所以高级语言是大多数编程者的首选。用高级语言编写的程序称为源程序，计算机系统不能直接理解和执行，必须将其转换为计算机系统能够认识、理解的机器语言才能被系统执行。通过一个示例来理解高级语言：

```python
操作:使用Python语言让计算机显示"Hello World"
 
print("Hello World")   高级语言
```
